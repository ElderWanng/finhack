```python
kline=AStock.getStockDailyPriceByCode('002624.SZ')
kline=kline[kline['trade_date']>'20210101']
kline['trade_date'] = pd.to_datetime(kline['trade_date'], format='%Y%m%d')
kline=kline.set_index('trade_date')   
 
print(kline)
```

```python
import mplfinance as mpf
import talib as ta

mycolor = mpf.make_marketcolors(up='red',down='green',inherit=True)
mystyle = mpf.make_mpf_style(gridaxis='both',gridstyle='-.',y_on_right=False,marketcolors=mycolor)
kline["macd"], kline["macd_signal"], kline["macd_hist"] = ta.MACD(kline['close'])
colors = ['r' if v >= 0 else 'g' for v in kline["macd_hist"]]
macd_plot = mpf.make_addplot(kline["macd"], panel=1,width=0.5,color='b')
macd_hist_plot = mpf.make_addplot(kline["macd_hist"], type='bar', panel=1, color=colors) 
plots = [macd_plot,macd_hist_plot]
mpf.plot(
         kline,
         type='candle',
         datetime_format='%Y-%m-%d',
         figratio=(20,10), 
         style=mystyle, 
         addplot=plots, 
         title=f"\n002624.SZ", 
         volume=True, 
         volume_panel=2,
         ylabel='', 
         ylabel_lower=''
)
```
