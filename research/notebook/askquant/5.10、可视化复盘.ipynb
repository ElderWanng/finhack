```python
#行业list
industry_pe_list=factorManager.getFactors(factor_list=['pe_0','peTtm_0'])

print(industry_pe_list)
```

```python
trade_date='20230116'
sql_stock_basic="select * from astock_basic"
df_stock_basic=mydb.selectToDf(sql_stock_basic,'tushare')
pe_list=[]
for row in df_stock_basic.itertuples():
    industry=getattr(row,"industry")
    ts_code=getattr(row,"ts_code")
    
    if ts_code not in industry_pe_list.index:
        break
    
    price_tmp=industry_pe_list.loc[ts_code]
    price_tmp['pe_rank'] = price_tmp['pe_0'].rank(pct = True)
    price_tmp['pe_ttm_rank'] = price_tmp['peTtm_0'].rank(pct = True)
    
    pe_now=price_tmp.loc[trade_date]['pe_0']
    pe_ttm_now=price_tmp.loc[trade_date]['peTtm_0']
    pe_max=price_tmp['pe_0'].max()
    pe_min=price_tmp['pe_0'].min()
#     print(pe_max)
#     print(pe_min)
#     print(pe_now)    
    pe_rank=price_tmp.loc[trade_date]['pe_rank']
    pe_ttm_rank=price_tmp.loc[trade_date]['pe_ttm_rank']    
    pe_info={
        "ts_code":ts_code,
        "industry":industry,
        "pe_rank":pe_rank,
        "pe_ttm_rank":pe_ttm_rank,
        "pe_now":pe_now,
        "pe_max":pe_max,
        "pe_min":pe_min,
        "pe_ttm_now":pe_ttm_now
    }
    
    pe_list.append(pe_info)
df_pe_list=pd.DataFrame(pe_list)

df_pe_median=df_pe_list.groupby('industry').median(numeric_only=True)
df_pe_median=df_pe_median.reset_index()


fig, ax = plt.subplots(figsize=(30,30))

for row in df_pe_median.itertuples():
    industry=getattr(row,"industry")
    pe_rank=getattr(row,"pe_rank")
    pe_ttm_rank=getattr(row,"pe_ttm_rank")
    
    ax.plot(pe_rank, pe_ttm_rank, label='pe/pe_ttm', c='r', ls='-.', marker='D', lw=2)
    ax.annotate(text=industry, xy=(pe_rank, pe_ttm_rank), xytext=(pe_rank, pe_ttm_rank+0.001),fontsize=14)    


plt.title('行业pe/pe_ttm历史分位',fontsize=32)
plt.xlabel('pe历史分位',fontsize=24)
plt.ylabel('pe_ttm历史分位',fontsize=24)
plt.grid(True)

plt.show()   
```
